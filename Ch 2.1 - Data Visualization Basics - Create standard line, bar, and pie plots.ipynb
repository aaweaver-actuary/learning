## 2.1 Create standard line, bar, and pie plots

```python
import matplotlib.pyplot as plt, seaborn as sb, numpy as np, pandas as pd
from numpy.random import randn
from matplotlib import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

### create line chart from a list obj - `plt.plot(x, y)`

```python
x = range(1, 10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x, y)
```

### create line chart from pandas - `Series.plot()`

```python
address = "./Ch01/01_05/mtcars.csv"
cars = pd.read_csv(address)
cars.columns = 'car_names mpg cyl disp hp drat wt qsec vs am gear carb'.split()
mpg = cars['mpg']
mpg.plot()
```

### line chart from data frame - `df.plot()`

```python
cars[['cyl', 'wt', 'mpg']].plot()
```

### bar chart from a list - `plt.bar(x,y)`

```python
plt.bar(x,y)
```

### bar chart from series - `Series.plot(kind='bar')`

```python
mpg.plot(kind='bar')
```

### horizontal bar chart from series - `Series.plot(kind='barh')`

```python
mpg.plot(kind='barh')
```

### pie chart - `plt.pie(); plt.show()`

```python
x = [1,2,3,4,0.5]
plt.pie(x)
plt.show()
```

### save a plot as an image file - `plt.savefig(filename)`

```python
plt.savefig('pie_chart.jpg')
plt.show()
```
