## 2.2 Define plot elements

### two methods for plot building
1. **functional method** - build plots by calling plotting function on varaible/set of variables  
2. **object-oriented method** - build plot by first generating a blank 'figure' object and the populating the object with plots/plot elements  


*talked about method1 in last section*

### steps on oop  
1. create blank figure object  
2. add axes to the figure  
3. generate plot(s) within the figure object  
4. specify plotting and layout parameters for the plots  

**subplot** - a plotting figure that contains more than one plot (or subplots)  
    * use `plt.subplots()`

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from numpy.random import randn
%matplotlib inline
plt.rcParams['figure.figsize'] = 5 , 4
```

```python
x = range(1, 10)
y = [1,2,3,4,0,4,3,2,1]
```

### blank figure -> axis -> plot
* axis - `[top, bottom, width, height]`

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x,y)
```

### set axis limits and tick marks
* need to define `plt.figure()` and `fig.add_axes([])` each time  

#### set x, y limits and tick marks with 
* `ax.set_xlim([start, end])`  
* `ax.set_ylim([start, end])`  
* `ax.set_xticks([...])` - with a list of where you want the tick marks  
* `ax.set_yticks([...])` - with a list of where you want the tick marks  

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

#x axis limits
ax.set_xlim([1,9])

#y axis limits
ax.set_ylim([0,5])

#tick marks
ax.set_xticks([0, 1, 2, 4, 5, 6, 8, 9, 10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

#### add in a grid with `ax.grid()`

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

#add a grid 
ax.grid()

ax.plot(x,y)
```

### for subplots, each subplot gets a figure & axis

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```
