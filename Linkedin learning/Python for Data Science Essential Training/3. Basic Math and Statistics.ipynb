## 3.4 Parametric methods

### Pearson Correlation: Assumptions
1. data normally distributed
2. continuous, numeric variables
3. variables linearly related

```python
import numpy as np, pandas as pd, scipy as sp, scipy.stats.stats as sps, matplotlib.pyplot as plt, seaborn as sb
plt.style.use('seaborn-whitegrid')
```

```python
## The Pearson correlation

cars = pd.read_csv('./Ch03/03_04/mtcars.csv')
n = cars.columns.to_list()
n[0] = 'car_name'
cars.columns = n
```

```python
sb.pairplot(cars); plt.show()
```

```python
x = cars['mpg hp qsec wt'.split()] 
sb.pairplot(x); plt.show()
```

```python
# use scipy to calculate pearson correlation
pearson_r_coef, p_value = sps.pearsonr(cars.mpg, cars.hp)
print('MPG vs HP')
print('Pearson R Correlation Coefficient: %0.3f' % pearson_r_coef)
```

```python
pearson_r_coef, p_value = sps.pearsonr(cars.mpg, cars.qsec)
print('MPG vs Qsec')
print('Pearson R Correlation Coefficient: %0.3f' % pearson_r_coef)
```

```python
pearson_r_coef, p_value = sps.pearsonr(cars.qsec, cars.hp)
print('Qsec vs HP')
print('Pearson R Correlation Coefficient: %0.3f' % pearson_r_coef)
```

### can also do this automatically with pandas

## 3.5 Nonparametric Methods for Correlation

1. Spearman's Rank Correlation
    * variables ordinal/nonlinear/nonnormal
2. Chi-square tables 
    * p < 0.05 => variables correlated
    * p > 0.05 => variables not correlated
    * need to bin variables

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, scipy as sp, scipy.stats as sps, seaborn as sb, matplotlib
matplotlib.rcParams['figure.figsize'] = 14,7
plt.style.use('seaborn-whitegrid')
```

```python
# spearman rank correlation
sb.pairplot(cars); plt.show()
```

```python
X = cars['cyl vs am gear'.split()]
sb.pairplot(X); plt.show()
```

1. Are these variables **ordinal**?
    * YES
2. Are these variables related **nonlinearly**?
    * YES -- what linear relationship is there?
3. **nonnormal**?
    * YES

```python
cyl, vs, am, gear = cars.cyl, cars.vs, cars.am, cars.gear
coef, pvalue = sps.spearmanr(cyl, vs)
print('Spearman rank coef: %0.3f' % coef)
```

```python
coef, pvalue = sps.spearmanr(cyl, am)
print('Spearman rank coef: %0.3f' % coef)
```

```python
coef, pvalue = sps.spearmanr(cyl, gear)
print('Spearman rank coef: %0.3f' % coef)
```

### chi square test for independence

```python
table = pd.crosstab(cyl, am)
chi2, p, dof, ev = sps.chi2_contingency(table.values)
print('Chi-Square Stat: {0:0.3f}, p-value: {1:0.3f}, dof: {2}'.format(chi2, p, dof))
```

```python
table = pd.crosstab(cyl, vs)
chi2, p, dof, ev = sps.chi2_contingency(table.values)
print('Chi-Square Stat: {0:0.3f}, p-value: {1:0.3f}, dof: {2}'.format(chi2, p, dof))
```

```python
table = pd.crosstab(cyl, gear)
chi2, p, dof, ev = sps.chi2_contingency(table.values)
print('Chi-Square Stat: {0:0.3f}, p-value: {1:0.3f}, dof: {2}'.format(chi2, p, dof))
```

## 3.6 Transform dataset distributions

### two ways to scale:
1. normalization
2. standardization

### when to use `scikit-learn` preprocessing

```python
matplotlib.rcParams.update({
    'xtick.labelsize':'x-large', 
    'ytick.labelsize':'x-large',
    'axes.titlesize':'large', 
    'figure.figsize': (15,8)
})
```

```python
import numpy as np, pandas as pd, scipy as sp, matplotlib.pyplot as plt, seaborn as sb, sklearn, matplotlib
from sklearn import preprocessing as skp
from sklearn.preprocessing import scale
```

```python
matplotlib.rcParams['figure.figsize'] = 15, 12; sb.set_style('whitegrid')
```

```python
mpg = cars.mpg
plt.plot(mpg); plt.show()
```

```python
pd.DataFrame(cars['mpg'.split()].describe())
```

```python
#scale mpg variable between 0 and 1

mpg_matrix = cars[['mpg']].to_numpy().reshape(-1, 1)
scaled = skp.MinMaxScaler()
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg); plt.show()
```

```python
# use a range other than 0 to 1
mpg_matrix = cars[['mpg']].to_numpy().reshape(-1, 1)
scaled = skp.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg); plt.show()
```

### use `.scale()` to scale features

```python
# turn off scaling to the mean/std deviation:

from sklearn.preprocessing import scale
standardized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
plt.plot(standardized_mpg); plt.show()
```

```python
# instead use the mean/sd ftn:
standardized_mpg = scale(mpg)
plt.plot(standardized_mpg); plt.show()

# this is standardized -- mean 0/sd 1
```
