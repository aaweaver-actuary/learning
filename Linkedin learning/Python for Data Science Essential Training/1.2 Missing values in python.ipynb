## 1.2 Missing values in python

* by default, missing values are `NaN`

#### will look at:
1. discovering what is missing
2. filling in for missing values  
3. counting missing values  
4. filter out missing values

### figuring out what is missing

```python
import numpy as np, pandas as pd
```

```python
missing = np.nan
rows = ['row1', 'row2', missing, 'row4', 'row5', 'row6', missing, 'row8']
series_obj = pd.Series(rows)
series_obj
```

#### Find missing values with `pd.Series.isnull()`

```python
series_obj.isnull()
```

### fill in missing values with `pd.DataFrame.fillna()`

```python
np.random.seed(25)
df_obj = pd.DataFrame(np.random.randn(36).reshape((6,6)))
df_obj.iloc[3:5, 0] = missing
df_obj.iloc[1:4, 5] = missing
df_obj
```

#### fill with 0's

```python
df_obj.fillna(0)
```

#### fill using a dictionary

```python
#pass a dict -- eg fill different numbers for different columns
df_obj.fillna({0: 0.1, 5:1.25})
```

#### forward fill down the column - `pd.DataFrame.fillna(method='ffill')`

```python
df_obj.fillna(method='ffill')
```

### count missing values

* eg which variables have most missings?

```python
df_obj
```

#### use `pd.DataFrame.isnull().sum()`

```python
df_obj.isnull().sum()
```

### filtering out missing values



```python
df_obj
```

#### drop every row with any missing values - `pd.DataFrame.dropna()`

```python
df_obj.dropna()
```

#### drop any column with missing values - `pd.DataFrame.dropna(axis=1)`

```python
df_obj.dropna(axis=1)
```

#### only drop rows that have all missing values - `pd.DataFrame.dropna(how='all')`

```python
df_obj.dropna(how='all')
```
